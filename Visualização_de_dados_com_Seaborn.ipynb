O Seaborn é uma biblioteca de visualização de dados do Python baseado no Matplotlib. Ele fornece uma interface de alto nível para construção de gráficos estatísticos atrativos e informativos

```python
import seaborn as sns

sns.set_theme(style = 'darkgrid') #define tema dos gráficos
Base_Dados = sns.load_dataset('tips') #carrega base de dados do seaborn
Base_Dados.head()
```

```python
Base_Dados.rename( columns = {'total_bill': 'Total Conta',
                              'tip': 'Gorjeta',
                              'sex': 'Sexo',
                              'smoker': 'Fumante',
                              'day': 'Dia Semana',
                              'time': 'Periodo',
                              'size': 'Pessoas na Mesa'}, inplace = True)
Base_Dados.head()
```

Gráfico relplot:

```python
sns.relplot(x = 'Total Conta', y = 'Gorjeta', data = Base_Dados)
```

```python
sns.relplot(x = 'Total Conta', y = 'Gorjeta', data = Base_Dados, hue = 'Sexo')
```

Gráfico de linhas:

```python
sns.relplot(x = 'Total Conta', y = 'Gorjeta', data = Base_Dados, kind = 'line')
```

```python
sns.relplot(x = 'Total Conta', y = 'Gorjeta', data = Base_Dados, kind = 'line', hue = 'Fumante')
```

Gráfico histplot:

```python
sns.histplot(data = Base_Dados, x = 'Total Conta')
```

```python
sns.histplot( data = Base_Dados, x = 'Total Conta', hue = 'Fumante')
```

Gráfico pairplot:

```python
sns.pairplot(Base_Dados)
```

```python
sns.pairplot(Base_Dados, hue = 'Sexo')
```

```python
sns.pairplot(Base_Dados, kind = 'kde')
```

Gráfico boxplot:

```python
sns.boxplot(data = Base_Dados, x ='Sexo', y = 'Total Conta', hue = 'Fumante')
```

```python
sns.boxplot(data = Base_Dados, x = 'Dia Semana', y = 'Total Conta')
```

```python
sns.boxplot(data = Base_Dados, x = 'Dia Semana', y = 'Total Conta', hue = 'Sexo')
```

Gráfico violinplot:

```python
sns.violinplot( data=Base_Dados, x='Dia Semana', y='Total Conta' )
```

Gráfico swarplot:

```python
sns.swarmplot(data=Base_Dados, x='Total Conta', y='Sexo' )
```

Gráfico jointplot:

```python
sns.jointplot( data=Base_Dados, x='Total Conta', y='Gorjeta')
```
